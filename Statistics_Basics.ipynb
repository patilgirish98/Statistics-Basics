  4. What is a box plot, and what can it tell you about the distribution of data?
      - Box Plot (Box-and-Whisker Plot)
      - A box plot is a visual tool that shows the spread, central tendency, and outliers of a dataset.
      - Key Components:-
         - Minimum & Maximum - Smallest and largest values within the range.
         - Q1 (25th percentile) - Lower quartile.
         - Median (Q2, 50th percentile) - Middle value.
         - Q3 (75th percentile) - Upper quartile.
         - Outliers - Extreme values beyond the whiskers.
      - Use Cases:-
         - Comparing datasets (e.g., exam scores).
         - Detecting outliers (e.g., income analysis)
         - Understanding data distribution in business and research.


  5. Discuss the role of random sampling in making inferences about populations.
     - Role of Random Sampling in Inferences:-
        - Random sampling selects a subset from a population where each member has an equal chance of being chosen. This helps in making accurate, unbiased inferences about the whole population.
     - Why It Matters:-
        - Reduces Bias - Ensures fair representation.
        - Supports Generalization - Sample results apply to the whole population.
        - Enables Reliable Statistical Analysis - Helps in hypothesis testing and predictions.

    - Types of Random Sampling:
        - Simple Random Sampling - Equal chance for all (e.g., lottery draw).
        - Stratified Sampling  Divides population into groups, then samples from each

 6.  Explain the concept of skewness and its types. How does skewness affect the interpretation of data?
    - Skewness & Its Impact :-
       - Skewness measures asymmetry in data distribution.
    - Types of Skewness:
       - Positive Skew (Right-Skewed)  - Tail on the right, Mean > Median > Mode (e.g., income distribution).
       - Negative Skew (Left-Skewed)  - Tail on the left, Mean < Median < Mode (e.g., exam scores).
       - Zero Skew (Symmetrical)  - Balanced, Mean = Median = Mode (e.g., heights of adults).

        - IQR = Q3 - Q1 (middle 50% of data).

        -  Not affected by extreme values (unlike mean and standard deviation).
        - Reliable for skewed distributions (better than range).
        - Used in box plots to visualize data spread and outliers.

 8.Conditions for Using the Binomial Distribution Discuss the conditions under which the binomial distribution is used.

    - Conditions for Using the Binomial Distribution
       - The binomial distribution is used when an experiment meets the following conditions:

    1.  Fixed Number of Trials (n)
      - The experiment is repeated a set number of times.
      - Example: Flipping a coin 10 times.

   2. Two Possible Outcomes (Success/Failure)
      - Each trial has only two outcomes
      - Example: A basketball shot is either made (success) or missed (failure).

   3. Constant Probability (p)
      - The probability of success remains the same for each trial.
      - Example: A fair die has a 1/6 chance of rolling a 6 each time.

   4. Independent Trials
      - The outcome of one trial does not affect the others.
      - Example: Each coin flip is independent of the previous one.
